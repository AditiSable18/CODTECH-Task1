```python
df = df[['ID', 'State', 'Candidate', 'Party', 'Total Votes', '% of Votes', 'Result']].copy()
```

```python
newdf = df.dropna(how="any").reset_index(drop=True)
newdf
```

```python
df_state = newdf[['State','Total Votes']]
df_state
```

```python
agg_func={'Total Votes':'sum',}
df_state = df_state.groupby(df_state['State']).aggregate(agg_func).reset_index()
df_state
```

```python
df_state = df_state.sort_values('Total Votes',ascending=False).head(5)
df_state
```

```python
mplt.hist(df_state['State'])
mplt.xlabel("State")
mplt.ylabel("Total Votes")
mplt.show()
```

```python
df_party = newdf[['Party','Total Votes']]
df_party
```

```python
agg_func={'Total Votes':'sum',}
df_party = df_party.groupby(df_party['Party']).aggregate(agg_func).reset_index()
df_party
```

```python
df_party= df_party.sort_values('Total Votes',ascending=False).head(12)
df_party
```

```python
df_party.plot(kind = 'scatter', y='Party', x='Total Votes')
```
